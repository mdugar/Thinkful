# 3.3.4 DRILL - Descriptive Statistics and Normality

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

#### 1. Random distribution: noncentral_chisquare (degrees of freedom, non-centrality, size=None)

```python
noncentral_chisquare = np.random.noncentral_chisquare(10, 2, 100)

plt.hist(noncentral_chisquare, bins=20)
plt.axvline(noncentral_chisquare.mean(), linestyle='solid', linewidth=2, color='red')
plt.axvline(noncentral_chisquare.mean() + noncentral_chisquare.std(), linestyle='dashed', linewidth=2, color='red')
plt.axvline(noncentral_chisquare.mean() - noncentral_chisquare.std(), linestyle='dashed', linewidth=2, color='red')
plt.show()
```

#### 2. Random distribution: laplace (loc, scale, size=None)

```python
laplace = np.random.laplace(0, 1, 100)

plt.hist(laplace, bins=20)
plt.axvline(laplace.mean(), linestyle='solid', linewidth=2, color='red')
plt.axvline(laplace.mean() + laplace.std(), linestyle='dashed', linewidth=2, color='red')
plt.axvline(laplace.mean() - laplace.std(), linestyle='dashed', linewidth=2, color='red')
plt.show()
```

#### 3. Random distribution: standard_normal (size)

```python
standard_normal = np.random.standard_normal(100)

plt.hist(standard_normal, bins=20)
plt.axvline(standard_normal.mean(), linestyle='solid', linewidth=2, color='red')
plt.axvline(standard_normal.mean() + standard_normal.std(), linestyle='dashed', linewidth=2, color='red')
plt.axvline(standard_normal.mean() - standard_normal.std(), linestyle='dashed', linewidth=2, color='red')
plt.show()
```

#### 4. Random distribution: vonmises(mu, kappa, size)

```python
vonmises = np.random.vonmises(1, 1, 100)

plt.hist(vonmises, bins=20)
plt.axvline(vonmises.mean(), linestyle='solid', linewidth=2, color='red')
plt.axvline(vonmises.mean() + vonmises.std(), linestyle='dashed', linewidth=2, color='red')
plt.axvline(vonmises.mean() - vonmises.std(), linestyle='dashed', linewidth=2, color='red')
plt.show()
```

#### 5. Random distribution: zipf(distribution parameter, size)

```python
zipf = np.random.zipf(3, 100)

plt.hist(zipf, bins=20)
plt.axvline(zipf.mean(), linestyle='solid', linewidth=2, color='red')
plt.axvline(zipf.mean() + zipf.std(), linestyle='dashed', linewidth=2, color='red')
plt.axvline(zipf.mean() - zipf.std(), linestyle='dashed', linewidth=2, color='red')
plt.show()
```

#### 6. Random distribution: rayleigh(scale, size)

```python
rayleigh = np.random.rayleigh(3, 100)

plt.hist(rayleigh, bins=20)
plt.axvline(rayleigh.mean(), linestyle='solid', linewidth=2, color='red')
plt.axvline(rayleigh.mean() + rayleigh.std(), linestyle='dashed', linewidth=2, color='red')
plt.axvline(rayleigh.mean() - rayleigh.std(), linestyle='dashed', linewidth=2, color='red')
plt.show()
```

#### Evaluation:
    For the mean and standard deviation to give useful information about the datapoints, the data given should not be randomed or changed, so that's the first issue. From the six different distributions showed here, Laplace distribution gives the most useful mean and standar deviation information. I think that a larger set would have also contributed to a better description.

#### Adtional activity:

##### 1. Two normally-distributed variables, a with a mean of 5 and standard deviation of 0.5, and b with a mean of 10 and standard deviation of 1.

```python
a = np.random.normal(5, 0.5, 2000)
b = np.random.normal(10, 1, 2000)
```

##### 2. Adding a and b to create c

```python
c = a + b
```

##### 3. Plot a histogram of c, mean and standard deviation. Additionaly compute mean and standard deviation

```python
plt.hist(c, bins=20)

plt.axvline(c.mean(), linestyle='solid', linewidth=2, color='red')
plt.axvline(c.mean() + c.std(), linestyle='dashed', linewidth=2, color='red')
plt.axvline(c.mean() - c.std(), linestyle='dashed', linewidth=2, color='red')
plt.show()
print("Mean: ", c.mean())
print("Standard deviation: ", c.std())
```

##### 4. Evaluation
We can see that the mean of c is the sum of the means of a and b. However, the latter doesn't happen with the standard deviation. Having a large dataset (2000 datapoints), gives us the possibility of seeing a more normal distribution, something we didn't quite see in the first excercise of this assignment.
