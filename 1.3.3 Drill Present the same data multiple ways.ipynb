# Drill: Present the same data multiple ways

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load CSV

wine = pd.read_csv('winequality-red.csv')
for wine_type in wine:
    wine['type'] = 'Red'
```

```python
wine.head()
```

### Plotting one variable (volatile acidity) in four different ways:

```python
#1 Histogram

plt.hist(wine['volatile acidity'], bins=20, color='red', alpha=.5)
plt.xlabel('Volatile Acidity')
plt.title('Plot 1: Histogram of Volatile Acidity')
plt.show()
```

```python
#2 Boxplot

sns.set_style("whitegrid")
sns.boxplot(wine['volatile acidity'])
plt.title('Plot 2: Boxplot of Volatile Acidity')
sns.despine(offset=10)
plt.show()
```

```python
#3 Barplot

sns.set_style("darkgrid")
sns.barplot(y="volatile acidity", data=wine, palette="Dark2_r")
plt.title('Plot 3: Barplot of Volatile Acidity')
sns.despine(offset=10)
plt.show()
```

```python
#4 Pointplot

sns.set(style="whitegrid")
sns.factorplot(y='volatile acidity', data=wine, size=5, kind="point", palette="pastel", ci=95, dodge=True)
sns.despine(left=True)
plt.title('Plot 4: Pointplot of Volatile Acidity')
plt.show()
```

### Pros and cons for each plot

* __1. Histogram:__ 

Pros: Shows the normal distribution of acidity in the different samples.

Cons: Doesn't show an average value, doesn't show exactly the value for each sample.

* __2. Boxplot:__

Pros: Shows the median acidity volatility, and minimum and maximums.

Cons: There are too many isolated points in the maximum.

* __3. Barplot:__

Pros: Shows the maximum volatile acidity.

Cons: Doesn't show a distribution, a central tendency or a mean.

* __4. Pointpllot:__

Pros: Shows mean.

Cons: Doesn't show standard deviation, highs, lows, distribution.

### Two continuous variables (fixed acidity and alcohol) plotted in three different ways

```python
#Scatterplot
w = sns.lmplot(y='fixed acidity', x='alcohol', data=wine, fit_reg=False, scatter_kws={'alpha':0.4})
w.set_ylabels("Fixed Acidity")
w.set_xlabels("Alcohol")
plt.title('Plot 1: Scatterplot - Fixed acidity by alcohol')
plt.show()

#Scatterplot with regression line
w = sns.lmplot(y='fixed acidity', x='alcohol', data=wine, fit_reg=True, scatter_kws={'alpha':0.4})
w.set_ylabels("Fixed Acidity")
w.set_xlabels("Alcohol")
plt.title('Plot 2: Scatterplot with regression line - Fixed acidity by alcohol')
plt.show()

#Joint kernel density estimate
w = sns.jointplot(y='fixed acidity', x='alcohol', data=wine, kind="kde", size=7, space=0)
plt.title('                                                                     Plot 3: Joint kernel density estimate - Fixed acidity by alcohol')
plt.show()
```

### Pros and cons for each plot

* __1. Scatterplot:__
Pros: Shows highs and lows of fixed acidity and alcohol.

Cons: Doesn't show an exact relationship.

* __2. Scatterplot with regression line:__
Pros: Shows a trend in the relationship.

Cons: As with the scatterplot, it can give you an idea of highs and lows, but not very exact.

* __3. Joint Kernel density estimate:__
Pros: Clearly shows the central estimation when both variable are correlated, its mean and individual standard deviation.

Cons: It would be great to plot exact lines to determine the latter values.

### One continuous variable (total sulfur dioxide) and one categorical variable (type) plotted them six different ways

```python
# Histogram
plt.hist(wine['total sulfur dioxide'], bins=20, color='red',  alpha=.5, label='Red')
plt.xlabel('Total sulfur dioxide')
plt.title('Plot 1: Histogram - Total sulfur dioxide')
plt.show()

# Swarmplot scatterpltot
sns.set(style='whitegrid', color_codes=True)
sns.swarmplot(x='type', y='total sulfur dioxide', data=wine);
plt.title('Plot 2: Swarmplot - Total sulfur dioxide')
plt.show()

# Barplot
g=sns.factorplot(x='type', y='total sulfur dioxide', hue='type', data=wine, kind='bar', palette='CMRmap', ci=95)
plt.title('Plot 3: Barplot - Total sulfur dioxide for red wines')
plt.show()

# Boxplot
sns.boxplot(x='type', y='total sulfur dioxide', hue='type', data=wine)
plt.title('Plot 4: Boxplot - Total sulfur dioxide for red wines')
sns.despine(offset=5, trim=True)
plt.show()

# Violin pointplot
sns.set(style="whitegrid")
g = sns.factorplot(x='type', y='total sulfur dioxide', hue='type', data=wine, kind='violin', ci=95, dodge=True, join=False)
plt.title('Plot 5: Violin pointplot - Total sulfur dioxide for red wine')
plt.show()

# Stripplot scatterpltots
sns.set(style='whitegrid', color_codes=True)
sns.stripplot(x='type', y='total sulfur dioxide', data=wine);
plt.title('Plot 6: Strip plot - Total sulfur dioxide')
plt.show()
```

### Pros and cons for each plot

Note: Unfortunately this dataset didn't include a categorical variable, so I added a "Type" variable for this set of red wines. However, there are no other types of wine present, so the comparison beween continous and categorical variables, in this case is not of much worth. 

* __1. Histogram:__
Pros: Shows a clear distribution for the total sulfur dioxide present in the samples, with a greater number between 0 and 25.

Cons: Doesn't show an average value for the total sulfur dioxide.

* __2. Swarmplot:__
Pros: Shows that a great part of the popullation of the dataset has a total sulfur dioxide concentration beween 0 and 70 approx, in other words a distribution similar to the histogram.

Cons: There is a lack of mean value, and exact values for the distribution.

* __3. Barplot:__
Pros: I don't find positive facts for this plot.

Cons: Doesn't show mean, mode, median or a central tendency.

* __4. Boxplot:__
Pros: Shows a median, an an approximate standard deviation, highs and lows.

Cons: Not accurate for determining exact values, and there are too many points above the highs which make it more difficult to interpret.

* __5. Violin pointplot:__
Pros: Shows a clear distribution, a median, standard devation and highs and lows.

Cons: Visually it's interesting, however you can't determine exact values.

* __5. Strip plot:__
Pros: Shows highs and lows.

Cons: Doesn't show mean, mode, median or a central tendency.
